```python
import tkinter as tk
import requests

class HavaDurumuUygulamasi:
    def __init__(self, pencere):
        self.pencere = pencere
        self.pencere.title("Hava Durumu Uygulaması")

        # Şehir adı girmek için bir etiket ve giriş kutusu
        self.etiket = tk.Label(pencere, text="Şehir Adı:")
        self.etiket.pack()

        self.sehirGirisi = tk.Entry(pencere)
        self.sehirGirisi.pack()

        # Hava durumunu gösteren buton
        self.buton = tk.Button(pencere, text="Hava Durumunu Göster", command=self.havaDurumunuGoster)
        self.buton.pack()

        # Sonuçları göstermek için bir etiket
        self.sonucEtiketi = tk.Label(pencere, text="", fg="blue")
        self.sonucEtiketi.pack()

    # Hava durumunu gösteren fonksiyon
    def havaDurumunuGoster(self):
        sehir = self.sehirGirisi.get().strip()

        # Eğer şehir adı boşsa, bir uyarı göster
        if sehir == "":
            self.sonucEtiketi.config(text="Lütfen bir şehir adı girin!", fg="red")
            return

        try:
            # Hava durumu verilerini getiriyoruz
            havaDurumu = self.havaDurumunuGetir(sehir)
            if havaDurumu:
                # Hava durumu başarılı bir şekilde alındıysa, sonucu göster
                self.sonucEtiketi.config(text=f"{sehir} için Hava Durumu: {havaDurumu['description'].capitalize()}, Sıcaklık: {havaDurumu['temp']}°C", fg="blue")
            else:
                # Hava durumu alınamadıysa bir hata mesajı göster
                self.sonucEtiketi.config(text="Hava durumu bilgisi alınamadı!", fg="red")
        except Exception as e:
            # Eğer bir hata oluşursa, hata mesajını göster
            self.sonucEtiketi.config(text=f"Hata: {str(e)}", fg="red")

    # API'den hava durumu verilerini alan fonksiyon
    def havaDurumunuGetir(self, sehir):
        apiAnahtari = "dcbe55c1e3134568a5824acb8fc1ca1a"
        url = f"http://api.openweathermap.org/data/2.5/weather?q={sehir}&units=metric&appid={apiAnahtari}&lang=tr"

        for _ in range(3):
            yanit = requests.get(url)
            if yanit.status_code == 200:
                veri = yanit.json()
                havaDurumu = {
                    "description": veri["weather"][0]["description"],
                    "temp": veri["main"]["temp"]
                }
                return havaDurumu
            else:
                continue
        return None

# Uygulamanın çalışması için gereken kod
pencere = tk.Tk()
uygulama = HavaDurumuUygulamasi(pencere)
pencere.mainloop()
```
